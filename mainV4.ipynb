```python
import pandas as pd
import csv 
from cryptography.fernet import Fernet
from csv import writer
from datetime import datetime

def generate_key():
    """
    Generates a key and save it into a file
    """
    key = Fernet.generate_key()
    with open("secret.key", "wb") as key_file:
        key_file.write(key)
    return key

def load_key():
    """
    Load the previously generated key
    """
    return open("secret.key", "rb").read()

def enter_key():
    """
    Load the previously generated key
    """
    key = bytes(input("\nEnter Key: "), 'utf-8')
    return key

def encrypt_message(message):
    """
    Encrypts a message
    """
    generate_key()
    key = load_key()
    encoded_message = message.encode()
    f = Fernet(key)
    encrypted_message = f.encrypt(encoded_message)
    print("\nENCRYPTING MESSAGE........\n")
   
    date_time = datetime.now()
    timestamp = date_time.strftime("%d/%m/%Y-%H:%M")
    List = [timestamp, message, encrypted_message, key]
    with open('msg.csv', 'a') as f_object:
        writer_object = writer(f_object)
        writer_object.writerow(List)
        f_object.close()
    return encrypted_message
        
def decrypt_message(encrypted_message):
    """
    Decrypts an encrypted message
    """
    try:
        key = enter_key()
        f = Fernet(key)
        decrypted_message = f.decrypt(encrypted_message).decode()
        print("\nDecrypted Message:\n", decrypted_message)
        return decrypted_message
    except ValueError:
        print("\n!!!!!!!Your Input is incorrect.  TRY AGAIN LATER"!!!!!!!)
    
flag = True
while flag == True:
    e_d = ["e", "d"]
    answer = input("\nDo you want to encrypt[e] or decrypt[d] a message? e or d: ").lower()
    
    #response_enc = response_enc.lower()
    if answer not in e_d:
        answer = input("\n!!!Please answer question!!! e or d: ")

    # encrypt message    
    if answer == "e":
        message = input("ENTER message to ENCRYPT:\n")
        if message !="":
            encrypt_message(message)
            break
        else:
            message = input("Message to encrypt cannot be empty. Please type your message:\n")
            encrypt_message(message)
            break

        # decrypt message
    elif answer == "d":
        y_n =["y", "n"]
        message = input("\nDo you want to decrypt the latest message? y or n: ").lower()
        
        if message not in y_n:
            message = input("\n!!!Let me ask you again... \nDo you want to decrypt the latest message? y or n: ").lower()
        # latest encrypted message
        if message == "y":
            
            file = pd.read_csv('msg.csv', header=None, sep=',')
            #help: iloc[startRow:endRow:startColoumn,endColumn]
            df = pd.DataFrame(file.iloc[-1:,1].values)
            print("Last message encrypted was:", df)
            break
        # encrypted message were user need key + encryption
        if message == "n":
            message = bytes(input("\nYou will need the ENCRYPTED MESSAGE and the KEY.\nEnter message to DECRYPT:\n"), "utf-8")
            decrypt_message(message)
            break
        # random answer by user (not y nor n)
        else:
            print("\Ok ciao...")
            break

    # other
    else:
        print("you type something wrong... Bye bye.")
        break

flag = False
```
